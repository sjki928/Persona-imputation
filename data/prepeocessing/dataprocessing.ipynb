```python
import torch

def kl_divergence_batch(mean_p, cov_p, mean_q, cov_q):
    # cov_p, cov_q는 shape이 (batch_size, 13, 13)이어야 하고
    # mean_p, mean_q는 shape이 (batch_size, 13)이어야 합니다.
    
    cov_q_inv = torch.inverse(cov_q)  # (batch_size, 13, 13)
    det_cov_p = torch.det(cov_p)      # (batch_size,)
    det_cov_q = torch.det(cov_q)      # (batch_size,)
    
    # Trace term
    trace_term = torch.einsum('bij,bji->b', cov_q_inv, cov_p)  # (batch_size,)
    
    # Mean difference term
    mean_diff = mean_q - mean_p  # (batch_size, 13)
    mean_term = torch.einsum('bi,bij,bj->b', mean_diff, cov_q_inv, mean_diff)  # (batch_size,)
    
    # Log determinant term
    log_det_term = torch.log(det_cov_q / det_cov_p)  # (batch_size,)
    
    # Dimensionality
    k = mean_p.shape[1]
    
    # KL divergence
    kl_div = 0.5 * (trace_term + mean_term - k + log_det_term)  # (batch_size,)
    return kl_div

# Example usage with random batch of 13-dimensional means and covariances
batch_size = 5
mean_p = torch.rand(batch_size, 13)
mean_q = torch.rand(batch_size, 13)
cov_p = torch.eye(13).expand(batch_size, 13, 13)  # Example covariance matrix for P (identity matrix)
cov_q = torch.eye(13).expand(batch_size, 13, 13) * 1.5  # Example covariance matrix for Q (scaled identity matrix)

kl_div = kl_divergence_batch(mean_p, cov_p, mean_q, cov_q)
print("KL Divergence (Batch):", kl_div)
```

```python
from sklearn.model_selection import train_test_split

def build_dataframe(csv_path,embedding_path,test_size=0.2):
    
    dataset = pd.read_csv(csv_path)
    
    scaler = StandardScaler()
    
    train_name ,test_name = train_test_split(dataset.VISIT_AREA_NM.unique(),test_size=test_size)
    
    train_mask = dataset.VISIT_AREA_NM.isin(train_name)
    test_mask = dataset.VISIT_AREA_NM.isin(test_name)
    
    train = dataset[train_mask]
    test = dataset[test_mask]

    scaler.fit(train.loc[:,STD_COL])
    
    train.loc[:,STD_COL] = scaler.transform(train.loc[:,STD_COL])
    test.loc[:,STD_COL] = scaler.transform(test.loc[:,STD_COL])

    def _add_embedding(dataset,embedding_path):
        
        dataframe = dataset.loc[:,['VISIT_AREA_NM']+STD_COL].groupby('VISIT_AREA_NM').mean().fillna(0)
        dataframe_var = dataset.loc[:,['VISIT_AREA_NM']+STD_COL].groupby('VISIT_AREA_NM').var().fillna(1)
        
        dataframe['mean'] = dataframe.apply(lambda x: [x[col] for col in STD_COL],axis=1) + 1e-8
        dataframe['var'] = dataframe_var.apply(lambda x: [x[col] for col in STD_COL],axis=1) + 1e-8
        dataframe = dataframe[['mean','var']].reset_index()
        
        dataframe = pd.merge(dataframe,dataset.loc[:,['VISIT_AREA_NM','overview']].drop_duplicates(),how='left',on='VISIT_AREA_NM')
        
        li = dataframe.overview.tolist()
        model = SentenceTransformer(embedding_path)
        embeddings = model.encode(li)
        dataframe['textvec'] = list(embeddings)
        
        return dataframe
    
    train = _add_embedding(train,embedding_path)
    test = _add_embedding(test,embedding_path)

    return train, test, scaler
```
